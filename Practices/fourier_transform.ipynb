```python
import numpy as np
import matplotlib.pyplot as plt

# Thông số tín hiệu
Fs = 100  # Tần số lấy mẫu (Hz)
T = 1     # Thời gian quan sát (giây)
t = np.linspace(0, T, Fs, endpoint=False)  # Trục thời gian
f = 5     # Tần số của tín hiệu (Hz)

# Tín hiệu sin
x = np.sin(2 * np.pi * f * t)

# Biến đổi Fourier (DFT)
X_f = np.fft.fft(x)  # Tính DFT của tín hiệu
frequencies = np.fft.fftfreq(Fs, 1/Fs)  # Trục tần số

# Chỉ lấy một nửa phổ (do tính đối xứng)
half_range = Fs // 2
X_magnitude = np.abs(X_f[:half_range])  # Lấy biên độ phổ
frequencies_half = frequencies[:half_range]  # Lấy trục tần số dương

# Vẽ tín hiệu gốc
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(t, x, label="x(t) = sin(2π5t)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Original Signal")
plt.legend()
plt.grid()

# Vẽ Phổ biên độ
plt.subplot(1, 2, 2)
plt.stem(frequencies_half, X_magnitude)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("Fourier Transform (Magnitude Spectrum)")
plt.grid()

plt.tight_layout()
plt.show()
```
