```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define time array (from 0 to 7 seconds)
t = np.linspace(0,2*np.pi, 1000)
```

```python
'''
üìå √ù nghƒ©a c·ªßa c√°c tham s·ªë:
‚úî A,B: Bi√™n ƒë·ªô ban ƒë·∫ßu c·ªßa hai s√≥ng sin.
‚úî a,b: H·ªá s·ªë t·∫Øt d·∫ßn (c√†ng l·ªõn, t√≠n hi·ªáu c√†ng gi·∫£m nhanh).
‚úî œâ1, œâ2: T·∫ßn s·ªë g√≥c c·ªßa hai s√≥ng sin.
'''
# Damped sinusoidal function
A = 10 # Amplitude 1
B = 20 # Amplitude 2
a = 0.5 # Damping factor 1 (H·ªá s·ªë t·∫Øt d·∫ßn c·ªßa s√≥ng th·ª© nh·∫•t)
b= 1 # Damping factor 2 (H·ªá s·ªë t·∫Øt d·∫ßn c·ªßa s√≥ng th·ª© hai)
omega1 = 4*np.pi # Angular frequency 1 (T·∫ßn s·ªë g√≥c c·ªßa s√≥ng th·ª© nh·∫•t)
omega2 = 2*np.pi # Angular frequency 2 (T·∫ßn s·ªë g√≥c c·ªßa s√≥ng th·ª© hai)

'''
üìå Ph√¢n t√≠ch c√¥ng th·ª©c:

x(t)=Ae^‚àíat * sin(œâ1*t) + Be^‚àíbt * sin(œâ2*t)

‚úî e^‚àíat v√† e^-bt: L√†m bi√™n ƒë·ªô s√≥ng gi·∫£m d·∫ßn theo th·ªùi gian.
‚úî sin(œâ1*t) v√† sin(œâ2*t): Hai s√≥ng sin c√≥ t·∫ßn s·ªë kh√°c nhau.
‚úî T·ªïng h·ª£p hai t√≠n hi·ªáu: K·∫øt h·ª£p hai s√≥ng sin c√≥ t·∫Øt d·∫ßn ƒë·ªÉ t·∫°o th√†nh t√≠n hi·ªáu ph·ª©c t·∫°p h∆°n.

üí° N·∫øu kh√¥ng c√≥ e^‚àíat, t√≠n hi·ªáu ch·ªâ l√† s√≥ng sin ƒë∆°n thu·∫ßn!
'''
x_t = A * np.exp(-a * t) * np.sin(omega1 * t) + B * np.exp(-b*t) * np.sin(omega2 * t)
```

```python
plt.figure()
plt.plot(t, x_t, color='blue')
plt.xlabel(r'$t(s)$')
plt.ylabel(r'Signal $x(t)$')
plt.title('(a)')
plt.xlim([0,7])
plt.ylim([-15,25])
plt.grid(True)
```
