```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Time array (from 0 to 1 second, 1000 points)
t = np.linspace(0, 1, 1000)

'''
T√≠n hi·ªáu sin t·ªïng qu√°t ƒë∆∞·ª£c vi·∫øt d∆∞·ªõi d·∫°ng:
x(t)= Asin(2œÄft + œï)
üîπ A: Bi√™n ƒë·ªô (Amplitude) ‚Äì quy·∫øt ƒë·ªãnh ƒë·ªô cao c·ªßa s√≥ng.
üîπ f: T·∫ßn s·ªë (Frequency) ‚Äì s·ªë chu k·ª≥ l·∫∑p l·∫°i trong 1 gi√¢y.
üîπ t: Th·ªùi gian (Time).
üîπ œÜ: Pha (Phase) ‚Äì d·ªãch chuy·ªÉn s√≥ng theo th·ªùi gian.
'''

'''
First signal: standard sine wave
So s√°nh v·ªõi c√¥ng th·ª©c tr√™n:
A = 1 ‚Üí Bi√™n ƒë·ªô c·ªßa s√≥ng sin l√† 1 (m·∫∑c ƒë·ªãnh).
f = 2 Hz ‚Üí T√≠n hi·ªáu n√†y c√≥ t·∫ßn s·ªë 2 Hz (t·ª©c l√† 2 chu k·ª≥ m·ªói gi√¢y).
œÜ = 0 ‚Üí Kh√¥ng c√≥ pha ban ƒë·∫ßu, s√≥ng b·∫Øt ƒë·∫ßu t·ª´ 0.
t: M·∫£ng th·ªùi gian t·ª´ 0 ƒë·∫øn 1 gi√¢y.
'''
signal1 = np.sin(2*np.pi * 5*t) 
```

```python
'''
T√≠n hi·ªáu cos t·ªïng qu√°t ƒë∆∞·ª£c vi·∫øt d∆∞·ªõi d·∫°ng:
x(t)= Acos(2œÄft + œï)
üîπ A: Bi√™n ƒë·ªô (Amplitude) ‚Äì quy·∫øt ƒë·ªãnh ƒë·ªô cao c·ªßa s√≥ng.
üîπ f: T·∫ßn s·ªë (Frequency) ‚Äì s·ªë chu k·ª≥ l·∫∑p l·∫°i trong 1 gi√¢y.
üîπ t: Th·ªùi gian (Time).
üîπ œÜ: Pha (Phase) ‚Äì d·ªãch chuy·ªÉn s√≥ng theo th·ªùi gian.
'''

'''
First signal: standard sine wave
So s√°nh v·ªõi c√¥ng th·ª©c tr√™n:
A = 0.5 ‚Üí S√≥ng c√≥ bi√™n ƒë·ªô 0.5 (gi√° tr·ªã l·ªõn nh·∫•t l√† 0.5, nh·ªè nh·∫•t l√† -0.5).
f = 5 Hz ‚Üí S√≥ng c√≥ t·∫ßn s·ªë 5 Hz (5 chu k·ª≥ m·ªói gi√¢y).
œÜ = -œÄ/4 ‚Üí S√≥ng b·ªã d·ªãch pha tr·ªÖ œÄ/4 radian.

üìå T·∫°i sao c√≥ "-œÄ/4" (pha tr·ªÖ)?

Khi pha √¢m (-œÄ/4), s√≥ng b·ªã d·ªãch sang ph·∫£i tr√™n tr·ª•c th·ªùi gian.
ƒêi·ªÅu n√†y c√≥ nghƒ©a l√† t√≠n hi·ªáu b·∫Øt ƒë·∫ßu ch·∫≠m h∆°n so v·ªõi m·ªôt t√≠n hi·ªáu cos b√¨nh th∆∞·ªùng.
'''
signal2 = 0.5 * np.cos(2*np.pi * 5*t - np.pi/4)
```

```python
# Create the plot
plt.figure()
plt.plot(t, signal1, label='Signal 1')
plt.plot(t, signal2, 'k--', label='Signal 2') 
plt.title('Test plots of sinusoids')
plt.xlabel('Time (sec.)')
plt.ylabel('Signal amplitudes')
plt.grid(True)
plt.legend()
```
